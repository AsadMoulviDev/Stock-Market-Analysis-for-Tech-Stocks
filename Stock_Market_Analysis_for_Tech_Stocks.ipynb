## What is the change in stock's price over time?

```python
#Using pandas we canplot the stocks adjusted closing price
AAPL.history(period="ytd")['Close'].plot(legend = True, figsize=(12,5))
```

```python
#Using pandas once more to plot the total volume being traded over time
AAPL.history(period="ytd")['Volume'].plot(legend=True,figsize=(12,5))
```

## What was the moving average of the stocks?

```python
#using pandas we will create moving averages for 10, 20 and 50 day periods of time
df=AAPL.history(period="ytd")
ma_days = [10,20,50]

for ma in ma_days:
    column_name = "MA %s days" %(str(ma))
    df[column_name] = AAPL.history(period="ytd")['Close'].rolling(window=ma,center=False).mean()
```

```python
df.tail()
```

```python
df[['Close','MA 10 days','MA 20 days','MA 50 days']].plot(figsize=(12,5))
```

## What was the daily return average of a stock?

```python
#In order calculate daily return we can use the percentage change of the adjusted closing price

df['Daily Return'] = df['Close'].pct_change()

df['Daily Return'].tail()
```

```python
#Let us now plot the daily return
df['Daily Return'].plot(figsize=(12,5))
```

```python
df['Daily Return'].plot(kind='hist',figsize=(12,5))
```
